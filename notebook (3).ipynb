```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Start coding here! Use as many cells as you like
# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
netflix_df.head(5)
```

```python
# Check the last 5 column in the dataset
netflix_df.tail()
```

```python
# check the description of the dataset
netflix_df.info()
```

```python
# Check for duplicate rows
duplicate_rows = netflix_df.duplicated().sum()
print("The number of duplicate rows are: ", duplicate_rows)
```

```python
# Check for missing values
missing_value = netflix_df.isna().sum()
print("Missing Values are: \n\n", missing_value)
```

```python
# Handle missing values in each column with NaN
# Columns to filled 
fill_col = ['director', 'cast', 'country', 'date_added']
netflix_df[fill_col] = netflix_df[fill_col].fillna("NaN", inplace=True)
print(netflix_df[fill_col])
print(netflix_df.head(5))
```

```python
nan_value = netflix_df.isnull().sum()
nan_value
```

```python
# Filter the data for movies between 1990 - 1999
netflix_subset_movies = netflix_df[(netflix_df['type'] == 'Movie') & (netflix_df['release_year'].between(1990, 1999))]
netflix_subset_movies
```

```python
# Calculate the frequency of each movie
duration_count = netflix_subset_movies['duration'].value_counts()

# Most frequent Movie duration
duration = duration_count.idxmax()
print("The most frequent movie duration in the 1990s is: ",duration)
```

```python
# Visualize the distribution of duration for movie released in the 1990s
import seaborn as sns

# Plot the distribution of movie durations
plt.figure(figsize=(10, 6))
sns.histplot(data=netflix_subset_movies, x='duration', bins=20, kde=True)
plt.title('Distribution of Movie Durations (1990s)')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count the number of short movies in the 1990s
short_movies = netflix_subset_movies[netflix_subset_movies['duration'] < 90]
short_movie_count = short_movies.shape[0]
print("The number of short movies in the 1990s is: ", short_movie_count)
```
